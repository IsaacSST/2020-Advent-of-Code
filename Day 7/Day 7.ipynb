```python
rules = f.read().split('\n')
```

```python
rules= rules[:-1]
```

```python
rules
```

```python
clean = list(map(lambda x:x.replace(' bags','').replace(' bag','').replace('contain','').replace(',', ' ').replace('.','').split('  '),rules))
```

```python
clean
```

```python
masterdict = dict()
```

```python
masterdict
```

```python
for item in clean:
    itemdict = dict()
    for contents in item[1:]:
        itemdict[contents[2:]] = contents[0]
    masterdict[item[0]] = itemdict
```

```python
masterdict
```

```python
masterdict['shiny gold']
```

```python
cleanmasterdict = masterdict.copy()
```

```python
for item in masterdict.keys():
    if ' other' in masterdict[item].keys():
        del cleanmasterdict[item]
```

```python
cleanmasterdict
```

```python
def numbags(colour,datadict):
    if colour in datadict.keys():
        total = 0
        for subcolour in datadict[colour].keys():
            if subcolour in datadict.keys():
                total += int(datadict[colour][subcolour]) + (int(datadict[colour][subcolour]))*numbags(subcolour,datadict)
            else:
                total += int(datadict[colour][subcolour])
            
        return total
            
    else:
        return 1
```

```python
ans = numbags('shiny gold',cleanmasterdict)
```
