```python
import numpy as np
import pandas as pd
```

```python
east = 0
north = 0
```

```python
directions = ['E','S','W','N']
```

```python
orientation = [0,'E']
```

```python
f = open('input.txt','r')
```

```python
instructions = f.read().split()
instructions
```

```python
for inst in instructions:
    if inst[0] == 'R':
        orientation[0] = (orientation[0] + (int(inst[1:]) // 90)) % 4
        orientation[1] = directions[orientation[0]]
    elif inst[0] == 'L':
        orientation[0] = (orientation[0] - (int(inst[1:]) // 90)) % 4
        orientation[1] = directions[orientation[0]]
    else:
        if inst[0] == 'F':
            direction = orientation[1]
        else:
            direction = inst[0]
        
        if direction == 'N':
            north += int(inst[1:])
        elif direction == 'S':
            north -= int(inst[1:])
        elif direction == 'E':
            east += int(inst[1:])
        else:
            east -= int(inst[1:])
```

```python
ans = abs(north) + abs(east)
ans
```
