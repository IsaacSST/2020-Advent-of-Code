```python
import numpy as np
import pandas as pd
```

## Part 1

```python
f = open('input.txt','r')
```

```python
nvalids = 0
for p in f:
    
    valid = False
    Min = int(p.split('-')[0])
    Max = int(p.split('-')[1].split(' ')[0])
    Char = p.split(' ')[1][0]
    Password = p.split(' ')[2][:-1]
    
    charcount = 0
    for c in Password:
        if c == Char:
            charcount +=1
            
    if (charcount <= Max) and (charcount >= Min):
        valid = True
        nvalids += 1
            
    print(Min, Max, Char, Password, valid, '\n', sep = ' ')
    
    
```
