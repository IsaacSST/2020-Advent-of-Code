```python
import numpy as np
import pandas as pd
```

```python
f = open('input.txt','r')
```

```python
groups = f.read().split('\n\n')
```

```python
groups
```

```python
clean = list(map(lambda x: x.replace("\n", ""),groups))
```

```python
clean
```

```python
countsum = 0
for group in clean:
    alphaset = set()
    for char in group:
        alphaset.add(char)
    
    countsum += len(alphaset)
    
```

```python
countsum
```

## Part 2

```python
count = 0
for group in groups:
    peeps = group.split('\n')
    allset = set(peeps[0])
    for person in peeps[1:]:
        if person != '':
            allset = allset.intersection(set(person))
            
    count += len(allset)
    
```
