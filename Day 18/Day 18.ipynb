```python
problems = f.read().split('\n')
```

```python
problems = problems[:-1]
```

```python
def evaluate(string):
    ans = 1
    op = '*'
    i = 0
    while i < len(string):
        if string[i] == '(':
            evaltup = evaluate(string[i+1:])
            if op == '*':
                ans = ans*evaltup[0]
            else:
                ans = ans+evaltup[0]
                
            i += evaltup[1]
                
        elif string[i] == ')':
            break
        elif string[i] == '*':
            op = '*'
            i += 1
        elif string[i] == '+':
            op = '+'
            i += 1
        elif string[i].isnumeric():
            if op == '*':
                ans = ans*int(string[i])
            else:
                ans = ans+int(string[i])
            i += 1
        else:
            i += 1
                
        
    return (ans,i+2)            
```

```python
answers = list(map(lambda x:evaluate(x)[0],problems))
```
