```python
completed = False
tempinst = instructions.copy()
for instruction in order:
    
    if instructions[instruction][0] == 'jmp':
        tempinst[instruction][0] = 'nop'
    elif instructions[instruction][0] == 'nop':
        tempinst[instruction][0] = 'jmp'
        
    accumulator = 0
    executed = [0]*631
    looping = False
    i = 0
        
    while not looping:
        current = tempinst[i]
        executed[i] = 1
        if current[0] == 'nop':
            i += 1
        elif current[0] == 'acc':
            i += 1
            accumulator += int(current[1])
        elif current[0] == 'jmp':
            i += int(current[1])
            
        if i == 631:
            completed = True
            break
    
        if executed[i] == 1:
            looping = True
            
    if completed:
        break
    else:
        tempinst[instruction][0] = instructions[instruction][0]
    
        
```
