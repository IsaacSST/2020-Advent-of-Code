```python
f = open('input.txt','r')
```

```python
inputs = f.read().split('\n\n')
```

```python
rules = inputs[0].split('\n')
```

```python
messages = inputs[1].split()
```

```python
rulesdict = dict()
```

```python
for rule in rules:
    split = rule.split(': ')
    rulesdict[split[0]] = split[1]
```

```python
rulesdict
```

```python
def construct(rule,rulesdict):
    possibilities = []
    thisrule = rulesdict[rule]
    if thisrule == '"a"' or thisrule == '"b"':
        possibilities.append(thisrule[1])
    else:
        if '|' in thisrule:
            rulesplit = thisrule.split(' | ')
        else:
            rulesplit = [thisrule]
            
        for rule in rulesplit:
            rulesplit2 = rule.split(' ')
            if len(rulesplit2) == 2:
                poss1 = construct(rulesplit2[0],rulesdict)
                poss2 = construct(rulesplit2[1],rulesdict)
                for p1 in poss1:
                    for p2 in poss2:
                        possibilities.append(p1 + p2)
                
            else:
                possibilities = possibilities + construct(rulesplit2[0],rulesdict)
    
    return possibilities
    
    
```

```python
rulezerovalids = construct('0',rulesdict)
```

```python
rulezerovalids
```

```python
ans = 0

for message in messages:
    
    if message in rulezerovalids:
        ans += 1
        
    
```

```python
ans
```

```python
for message in messages:
    add = True
    if len(message) >= 32 and len(message) % 8 == 0:
        if message[0:8] in rule42valids and message[8:16] in rule42valids and message[-8:] in rule31valids:
            count42 = 0
            count31 = 0
            stop42 = False
            for i in range(0,(len(message) // 8)):
                if message[8*i:8*(i+1)] in rule42valids and not stop42:
                    count42 += 1
                elif message[8*i:8*(i+1)] in rule31valids:
                    stop42 = True
                    count31 += 1
                else:
                    add = False
                    break
                    
            if count31 >= count42:
                add = False
                
            
        else:
            add = False
    else:
        add = False
        
    if add:
        extravalids += 1
            
            
        
```
