## Part 1

```python
f = open('input.txt','r')
```

```python
seatids = []
```

```python
for code in f:
    rbounds = [0,127]
    cbounds = [0,7]
    for i in range(0,10):
        char = code[i]
        if char == 'F':
            rbounds[1] = rbounds[0] + ((1 + rbounds[1] - rbounds[0])//2) - 1
        elif char == 'B':
            rbounds[0] = rbounds[0] + ((1 + rbounds[1] - rbounds[0])//2)
        elif char == 'L':
            cbounds[1] = cbounds[0] + ((1 + cbounds[1] - cbounds[0])//2) - 1
        elif char == 'R':
            cbounds[0] = cbounds[0] + ((1 + cbounds[1] - cbounds[0])//2)
            
    print(rbounds,cbounds)
    seatids.append(8*rbounds[0] + cbounds[0])
    
        
        
```

```python
f.close()
```
