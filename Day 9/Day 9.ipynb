```python
def findfirst(sq):
    for i in range(25,len(sq) + 1):
        current = sq[i-25:i]
        if sq[i] >= 2*max(current):
            return sq[i]
        else:
            found = False
            for j in range(0,25):
                for k in range(j+1,25):
                    if current[j] + current[k] == sq[i]:
                        found = True
                        
            if not found:
                return sq[i]
        
        
    
```

```python
def findweakness(num,sq):
    for i in range(0,len(sq)-1):
        for j in range(i+2,len(sq) + 1):
            thissum = sum(sq[i:j])
            if thissum > num:
                break
            elif thissum == num:
                return min(sq[i:j]) + max(sq[i:j])
```
